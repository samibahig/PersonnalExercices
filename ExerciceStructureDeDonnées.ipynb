```
# Prend comme argument une chaîne de caractères
# Retourne chaque élément de la chaîne de caractères (sans les espaces) dans une liste
def stringToList(sequence):
    return sequence.split()

# Prend comme argument une chaîne de caractères en ordre POSTFIXE
# Retourne une chaîne de caractères en ordre INFIXE
def postfixToInfix(equation):
    stack = []
    listOfElements = stringToList(equation)
    for i in range(0,len(listOfElements)):
        if listOfElements[i].isalnum(): #Si l'élément est alphanumérique, alors c'est un opérande
            stack.append(listOfElements[i])
        elif listOfElements[i] == '!':
            binaryOperand = stack.pop(-1)  # -1 fait référence au dernier élément du stack
            stack.append("%s%s" % (binaryOperand, listOfElements[i]))
        elif listOfElements[i] == '√':
            binaryOperand = stack.pop(-1)
            stack.append("%s%s" % (listOfElements[i], binaryOperand))
        else: #Si l'élément est ni alphanumérique ni un op unaire, alors c'est un opérateur binaire
            rightOperand = stack.pop(-1)
            leftOperand = stack.pop(-1)
            stack.append("(%s%s%s)" % (leftOperand, listOfElements[i], rightOperand))
    result = stack[0]
    return result[1:-1] # Enlever la première et dernière parenthèse...elles sont inutiles

# Ceci est le main()
result = postfixToInfix("9 √ 3 =")
print(result)
```
