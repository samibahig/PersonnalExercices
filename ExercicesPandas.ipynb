```
import pandas as pd
df = pd.DataFrame({'A' : [1,2,6,8,-1], 'B' : [0.1, 0.2,0.2,10.1,None], 'C' : ['a','b',None,'c','a']})
df
```

```
autre = pd.DataFrame({'C' : ['a','b'], 'D' : [1, 2]})
autre
```

```
df[~df['C'].isin(autre['C'])]
```

```
pd.merge(df, autre, on='C', how='inner', indicator=True)
```

```
pd.merge(df, autre, left_index=True, right_index=True, how='outer', indicator=True)
```

```
pd.merge(df, autre, on='A', how='left', indicator=True)
```

```
pd.merge(df, autre, left_on='A', right_on='D', indicator=True)
```

```
pd.merge(df, autre, left_index=True, right_index=True, how='right', indicator=True)
```
